###Dataset Features (Columns Description)

| Column Name          | Explanation                                                                                    |
| -------------------- | ---------------------------------------------------------------------------------------------- |
| **coin_name**        | The full name of the cryptocurrency (e.g., Bitcoin, Ethereum).                                 |
| **symbol**           | The short code used to represent the coin (e.g., BTC, ETH).                                    |
| **market_cap_rank**  | The rank of the coin based on its market capitalization (1 = highest value coin).              |
| **current_price**    | The latest price of the cryptocurrency (usually in USD).                                       |
| **market_cap**       | Total value of the coin = current price × total circulating supply. Shows how big the coin is. |
| **total_volume**     | Total amount of the coin traded in the last 24 hours. Indicates trading activity.              |
| **price_change_24h** | The change in price over the last 24 hours (can be positive or negative).                      |
| **high_24h**         | The highest price the coin reached in the last 24 hours.                                       |
| **low_24h**          | The lowest price the coin reached in the last 24 hours.                                        |
| **category**         | Type or classification of the coin (e.g., DeFi, NFT, Gaming, Stablecoin).                      |
| **volatility**       | Measures how much the price fluctuates. High volatility = big price swings.                    |
| **trend**            | Overall direction of the price (e.g., upward, downward, or stable).                            |

## Stage 2 – Data Cleaning and Pre-processing

```python
df.drop(columns=['market_cap_rank'], inplace= True)
```

```python
df.columns=df.columns.str.title()
df
```

```python
df['Price_Change_24H']= df['Price_Change_24H'].fillna(df['Price_Change_24H'].median())
df['Price_Change_24H']
```

```python
df['High_24H'] = df['High_24H'].fillna(df['High_24H']. median())
```

```python
df['Low_24H']= df['Low_24H'].fillna(df['Low_24H']. median())
df['Low_24H']
```

```python
df['Volatility']= df['Volatility'].fillna(df['Volatility']. median())
df['Volatility']
```

```python
df.rename(columns={'coin_name':'Coin_Name'}, inplace=True)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df=df.drop_duplicates()
df.tail()
```

```python
df.skew(numeric_only=True)
```

```python
df.kurt(numeric_only=True)
```

```python
cols = ['Current_Price', 'Market_Cap', 'Total_Volume',
        'High_24H','Low_24H' , 'Volatility']
```

```python
for col in cols:
  df["log_" + col]=np.log1p(df[col])
```

```python
df = df.round(3)
```

## Stage 3 – EDA and Visualizations

```python
#Distribution of Cryptocurrency Prices - Univariate

plt.figure(figsize=(10,5))
sns.histplot(df['Current_Price'], bins=50, kde=True, color='blue')
plt.title('Distribution of Cryptocurrency Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

##Distribution of Cryptocurrency Prices

Features Used: Current_Price
Purpose: To analyze the distribution and skewness of crypto prices.

1.   The price distribution is highly right-skewed, with most cryptocurrencies priced at the lower end.
2. Only a few coins have extremely high prices, creating the long tail in the distribution.
3. This indicates that the market is dominated by many low-value altcoins and a
small number of premium assets.





```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Total_Volume'])
plt.title('Boxplot of 24h Trading Volume')
plt.show()
plt.show()
```

##Boxplot of 24h Trading Volume

Features Used:
Total_Volume

Purpose: To identify spread, median, and outliers in trading activity.

1. Most cryptocurrencies have low to moderate 24h trading volume, as shown by the compressed box near zero.
2. A small number of coins appear as extreme outliers with very high trading activity.
3. This suggests that market liquidity is concentrated in a few major cryptocurrencies, while most coins are less actively traded.

```python
#Price vs Market Cap - Bivariate

plt.figure(figsize=(10,6))
sns.scatterplot(x='Market_Cap', y='Current_Price', data=df)
plt.title('Price vs Market Cap')
plt.xlabel('Market Cap')
plt.ylabel('Current Price')
plt.show()
```

##Price vs Market Cap Scatterplot
Features Used: Current_Price,
Market_Cap

Purpose: To study the relationship between valuation and pricing.

1. The scatterplot shows that high market-cap coins are generally associated with higher prices, though the relationship is not perfectly linear.
2. A few outlier coins dominate both price and market capitalization, likely representing top assets like Bitcoin.
3. Most coins are clustered at the lower end, indicating smaller-cap assets with lower price levels and higher speculative risk.

```python
#Correlation Heatmap for Numerical Features

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## Correlation Heatmap for Numeric Columns

Features Used:
Current_Price,
Market_Cap,
Total_Volume,
Price_Change_24H,
High_24H,
Low_24H,
Volatility

Purpose: To identify linear relationships and price drivers.

1. Market Cap and Volatility show a strong positive relationship (~0.96), indicating large-cap coins experience high trading activity and active price movement.
2. Current Price is highly correlated with 24H High and 24H Low, confirming daily price ranges strongly reflect overall price positioning.
3. Market Cap and Total Volume have moderate correlation (~0.63), suggesting higher-cap coins generally attract more trading participation.
4. Price_Change_24H has weak correlation with other variables, meaning short-term price movement depends more on market sentiment than structural indicators

```python
##Correlation Matrix
corr = df[['High_24H', 'Low_24H', 'Volatility', 'Price_Change_24H']].corr()

### Heatmap plot
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```

###Correlation Matrix — Interpretation
1. High_24H and Low_24H show perfect positive correlation (~1.0), meaning coins with higher daily highs also maintain proportionally higher daily lows, indicating consistent price ranges.

2. Volatility has moderate positive correlation with High_24H and Low_24H (~0.57–0.58), suggesting coins with wider daily price ranges tend to experience stronger market movement.

3. Price_Change_24H shows almost no correlation with other variables, indicating short-term price change is largely sentiment-driven rather than range-dependent.
4. Daily price range (High–Low) is a strong indicator of volatility but not reliable for predicting next-day price direction.

```python
#Pairplot - Multivariate

sns.pairplot(df[['log_Current_Price', 'log_Market_Cap', 'log_Total_Volume', 'Price_Change_24H']])
plt.show()
```

##Pairplot - Multivariate

Feature Used: log_Current_Price,
log_Market_Cap,
log_Total_Volume,
Price_Change_24H.

Purpose: To analyze relationships among multiple financial indicators

1. Market Cap and Total Volume show a clear positive relationship, proving that larger cryptocurrencies attract higher market participation and liquidity.
2. The log-transformed variables reveal clearer clustering and linear relationships, making hidden patterns in skewed financial data easier to interpret.
3. Price_Change_24H remains highly dispersed, confirming that short-term volatility cannot be explained by market size or volume alone.

```python
#Grouped Analysis – Average Price by Market Cap Rank Tier

df['Market_Cap_Group'] = pd.qcut(df['Market_Cap'], 5)
avg_price_by_group = df.groupby('Market_Cap_Group')['Current_Price'].mean().reset_index()

plt.figure(figsize=(16,5))
sns.barplot(x='Market_Cap_Group', y='Current_Price', data=avg_price_by_group)

plt.title('Average Cryptocurrency Price by Market Cap Quintile')
plt.show()
```

## Average Cryptocurrency Price by Market Cap Quintile

Feature Used: Market_Cap and Currect_Price,      

Purpose: To compare average prices across market cap segments

1. Market Cap and Total Volume show a positive relationship, confirming larger coins attract higher trading activity.
2. Log-transformed variables reveal clearer clustering patterns, helping better understand relationships among skewed financial data.
3. Price_Change_24H is widely scattered across all variables, showing short-term movement is unpredictable using market size or volume alone.
4. Most cryptocurrencies cluster in lower price ranges, while a few high-value coins dominate the upper range of the market.

```python
pivot_table = pd.pivot_table(df,
                        values='Current_Price',
                        index='Coin_Name',
                        aggfunc='mean')
pivot1 =pivot_table.sort_values(by='Current_Price', ascending=False)
pivot1.head(10)
```

```python
pivot1.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Cryptocurrencies by Current Price")
plt.ylabel("Current Price")
plt.xlabel("Coin Name")
plt.show()
```

### Top 10 Cryptocurrencies by Current Price

1. Few cryptocurrencies have significantly higher average prices compared to others.
2. High price concentration in select assets.
3. Price alone does not determine market dominance, as some lower-priced coins may have higher market capitalization.

```python
pivot2 = pd.pivot_table(df,
                          values='Market_Cap',
                          index=['Coin_Name', 'Category'],
                          aggfunc='mean')
pivot2 = pivot2.sort_values(by='Market_Cap', ascending=False)

pivot2.head(10)
```

```python
pivot2.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Cryptocurrencies by Market Cap")
plt.ylabel("Market Cap")
plt.xlabel("Coin Name and category")
plt.xticks(rotation=25)
plt.show()
```

###Market Cap by Coin and Category

1. Bitcoin and Ethereum dominate the market in terms of total market capitalization
2. Strong investor trust and large-scale adoption compared to other cryptocurrencies
3. Certain categories such as DeFi, Store-of-Value, or Stablecoins contribute a larger share of the total market cap.
4. Investors tend to prefer specific segments of the crypto market.

```python
pivot3 = pd.pivot_table(df,
                        values='Current_Price',
                        index=['Category', 'Coin_Name'],
                        columns='Trend',
                        aggfunc='mean')

pivot3 = pivot3.sort_values(by='Up', ascending=False)

pivot3.head(10).plot(kind='bar', figsize=(12,6))

plt.title("Top 10 Cryptocurrencies by Trend (Based on Price)")
plt.ylabel("Average Price")
plt.xlabel("Category and Coin Name")
plt.xticks(rotation=65)

plt.show()
```

###Top 10 Cryptocurrencies by Trend (Based on Price)
1. Bitcoin clearly dominates the market with the highest average price, reinforcing its position as the leading large-cap cryptocurrency and primary market driver.
2. Most top-ranked cryptocurrencies in the chart show an upward trend, indicating positive short-term market momentum among major assets.
3. Mid-cap coins like Pax Gold and Tether Gold maintain stable high-value positioning, suggesting their role as asset-backed or value-preserving cryptocurrencies.
4. Smaller-cap cryptocurrencies (e.g., UCHAI, Monero, Bittensor) show lower price levels but upward movement, indicating potential growth opportunities with comparatively higher investment risk.
